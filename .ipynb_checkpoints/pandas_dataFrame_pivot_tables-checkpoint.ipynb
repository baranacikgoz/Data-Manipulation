```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.groupby("sex")["survived"].describe()
#"sex"e gÃ¶re grupla, "survived"=1(kurtuldu) olanlarÄ± describe et
```

```python
titanic.groupby("sex")[["survived"]].describe()#dataframe olsun istiyosak
#iki tane kÃ¶ÅŸeli parantez koyabiliriz.
```

```python
titanic.groupby(["sex", "class"])[["survived"]].describe()
#hem "sex"e hem "class"a gÃ¶re grupla ve Ã¼sttekinin aynÄ±sÄ±nÄ± yap
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean")
#hem "sex"e hem "class"a gÃ¶re grupla ve hayatta kalanlarÄ±n ortalamasÄ±nÄ± gÃ¶ster
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack()
#Ã¼sttekinin aynÄ±sÄ± ancak farklÄ± okuma olanaÄŸÄ±
```

```python
#pivot ile table
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class")
#Ã¼sttekilerin hepsini daha az Ã§aba ve kodla yaptÄ±k
```
